```python
import requests
start = input("Введите начальную дату через - : ")
end = input("Введите конечную дату через - : ")
latitude = input("Введите координаты по широте : ")
longitude = input("Введите координаты по длинне : ")
radius = input("Введите радиус : ")
minmag = input("Введите минимальную магнитуду : ")
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query?'
response = requests.get(url, headers={'Accept':'application/json'}, params={
        'format':'geojson',
        'starttime':start,
        'endtime':end,
        'latitude':latitude,    # координаты Лондона
        'longitude':longitude,   # координаты Лондона
        'maxradiuskm':radius,
        'minmagnitude':minmag

	})
#print(response.text)

data = response.json()
count = data['metadata']['count']
for i in range(count):
    place = data['features'][i]['properties']['place']
    mag = data['features'][i]['properties']['mag']
    print(f'{i+1}. Place: {place}. Magnitude: {mag}')
```

```python
import requests
url = 'http://jsonplaceholder.typicode.com/todos'
response = requests.get(url)
listId = [1]
countId =[]
counter = 0
count = 0

data = response.json()
for res in data:
    if res['userId'] == listId[counter]:
        count += 1
    else:
        countId.append(count)
        listId.append(res['userId'])
        counter += 1
        count = 0
        
maximal = max(countId)
id = 0
for i in range(len(countId)):
    if countId[i] == maximal:
        id = i+1

print(f'Айди пользователя {id}. Кол-во решенных задач {maximal}')
        
```

```python
import csv

results = []
with open('MallCustomers.csv') as File:
    reader = csv.DictReader(File)
    for row in reader:
        results.append(row)
average = 0
count = 0
for el in results:
    if el["Gender"] == "Male" and int(el["Age"]) > 30:
        average += int(el['Annual Income (k$)'])
        count+=1
average /= count
average
```
